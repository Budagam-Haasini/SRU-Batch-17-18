```python
import pandas as pd
import matplotlib.pyplot as plt
data = {"Age": [18,22,17,20,18],
        "BMI": [38.5,25.5,35.4,24.1,22.15]}
df=pd.DataFrame(data)
Age=df["Age"]
BMI=df["BMI"]
plt.plot(Age,BMI,color="blue",linestyle="--")
plt.title("healthe care dataset")
plt.xlabel("BMI")
plt.ylabel("Age")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {"Age": [18,22,17,20,18],
        "BMI": [38.5,25.5,35.4,24.1,22.15]}
df=pd.DataFrame(data)
Age=df["Age"]
BMI=df["BMI"]
plt.scatter(Age,BMI,color="red",label="BMI")
plt.title("healthe care dataset")
plt.xlabel("BMI")
plt.ylabel("Age")
plt.legend()
plt.show()
```
