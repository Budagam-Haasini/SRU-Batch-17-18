```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Rating': [7.8, 6.5, 8.2, 5.9, 9.0, 7.2, 8.8, 6.9, 5.4, 7.6]*10,
    'Budget': [120, 80, 200, 40, 300, 90, 150, 70, 60, 100]*10,
    'BoxOffice': [400, 250, 600, 100, 900, 350, 500, 180, 120, 450]*10,
    'Genre': ['Action', 'Comedy', 'Drama', 'Thriller', 'Action',
              'Romance', 'Comedy', 'Drama', 'Action', 'Thriller']*10
}
df = pd.DataFrame(data)
#histogram
plt.figure(figsize=(7,5))
sns.histplot(df['Rating'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
#boxplot
sns.boxplot(x='Genre', y='BoxOffice', data=df, palette='Set2')
plt.title('Box Office Collection Across Genres')
plt.xlabel('Genre')
plt.ylabel('Box Office Collection (in million $)')
plt.show()
#density plot
sns.kdeplot(df['Budget'], fill=True, color='coral')
plt.title('Density Plot of Movie Budgets')
plt.xlabel('Budget (in million $)')
plt.ylabel('Density')
plt.show()
#pairplot
sns.pairplot(df, vars=['Budget', 'BoxOffice', 'Rating'], hue='Genre', palette='husl')
plt.suptitle('Relationships among Budget, Box Office, and Ratings', y=1.02)
plt.show()
```

Plot Type	Insight
Histogram (Ratings)	Shows whether most movies have high, medium, or low ratings (e.g., skewed or normal).
Box Plot (Box Office by Genre)	Helps identify which genre generally earns more and the spread of earnings.
Density Plot (Budget)	Reveals how budgets are distributed (e.g., most films low-budget, few high-budget).
Pair Plot	Displays pairwise relationships — e.g., do higher budgets correlate with higher ratings or box office collections?
