## Data cleaning and preparation

### Subtask:
Clean and prepare each DataFrame for analysis and visualization. This may involve handling missing values, converting data types, and renaming columns.


**Reasoning**:
Clean and prepare the dataframes by converting relevant columns to numeric types, handling commas, and displaying the results.



```python
import numpy as np

def clean_numeric_column(df, column_name):
    """Cleans a column with commas and converts it to numeric."""
    if column_name in df.columns:
        df[column_name] = df[column_name].astype(str).str.replace(',', '', regex=False)
        # Attempt conversion, coercing errors to NaN
        df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
    return df

# Clean 'women' DataFrame
women_df = dataframes['women']
for col in ['Total Women MSME', 'Micro', 'Small', 'Medium', 'Employment', 'Investment (Cr)', 'Turnover (Cr)']:
    women_df = clean_numeric_column(women_df, col)
dataframes['women'] = women_df

# Clean 'skills2' DataFrame
skills2_df = dataframes['skills2']
for col in ['Total_Trained (Youth)', 'Total_Placed', 'Budget_Allocated_Cr']:
    skills2_df = clean_numeric_column(skills2_df, col)
dataframes['skills2'] = skills2_df

# Clean 'it companies' DataFrame
it_companies_df = dataframes['it companies']
# Extract numeric part from 'Revenue (INR Cr)'
it_companies_df['Revenue (INR Cr)'] = it_companies_df['Revenue (INR Cr)'].astype(str).str.replace(',', '', regex=False)
it_companies_df['Revenue (INR Cr)'] = it_companies_df['Revenue (INR Cr)'].str.extract(r'(\d+\.?\d*)').astype(float)
dataframes['it companies'] = it_companies_df


# Check for missing values after conversion
print("Missing values after conversion:")
for df_name, df in dataframes.items():
    print(f"\n--- Missing values for DataFrame: {df_name} ---")
    print(df.isnull().sum())

# No column renaming is explicitly necessary based on the current column names and task.

# Display info and head of cleaned dataframes
for df_name, df in dataframes.items():
    print(f"\n--- Info for Cleaned DataFrame: {df_name} ---")
    df.info()
    print(f"\n--- Head for Cleaned DataFrame: {df_name} ---")
    display(df.head())
```

## Create visualizations

### Subtask:
Generate visualizations based on the analysis goals identified in the `analysis_goals` dictionary, excluding the cross-dataset analysis visualizations for now.


**Reasoning**:
Generate the requested visualizations for each analysis goal, excluding the cross-dataset analysis, based on the instructions and the analysis_goals dictionary.



```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# 1. India's Macroeconomic Trends
india_growth_df = dataframes['india_growth']

# Line plot of GDP over the years
fig_gdp = px.line(india_growth_df, x='Year', y='GDP (Trillion USD)', title='India\'s GDP Over the Years')
fig_gdp.update_layout(xaxis_title='Year', yaxis_title='GDP (Trillion USD)')
fig_gdp.show()

# Line plot of Growth Rate over the years
fig_growth_rate = px.line(india_growth_df, x='Year', y='GrowthRate (%)', title='India\'s Growth Rate Over the Years')
fig_growth_rate.update_layout(xaxis_title='Year', yaxis_title='Growth Rate (%)')
fig_growth_rate.show()

# Line plot of Unemployment Rate over the years
fig_unemployment_rate = px.line(india_growth_df, x='Year', y='UnemploymentRate (%)', title='India\'s Unemployment Rate Over the Years')
fig_unemployment_rate.update_layout(xaxis_title='Year', yaxis_title='Unemployment Rate (%)')
fig_unemployment_rate.show()

# Combined line plot of Growth Rate and Unemployment Rate over the years
fig_combined_macro = make_subplots(specs=[[{"secondary_y": True}]])

fig_combined_macro.add_trace(go.Scatter(x=india_growth_df['Year'], y=india_growth_df['GrowthRate (%)'], name='Growth Rate (%)'), secondary_y=False)
fig_combined_macro.add_trace(go.Scatter(x=india_growth_df['Year'], y=india_growth_df['UnemploymentRate (%)'], name='Unemployment Rate (%)'), secondary_y=True)

fig_combined_macro.update_layout(title_text='Growth Rate vs. Unemployment Rate Over the Years')
fig_combined_macro.update_xaxes(title_text='Year')
fig_combined_macro.update_yaxes(title_text='Growth Rate (%)', secondary_y=False)
fig_combined_macro.update_yaxes(title_text='Unemployment Rate (%)', secondary_y=True)
fig_combined_macro.show()


# 2. Women-led MSMEs Analysis
women_df = dataframes['women']

# Bar plot of 'Total Women MSME' by 'State'
fig_total_msme = px.bar(women_df, x='State', y='Total Women MSME', title='Total Women-led MSMEs by State')
fig_total_msme.update_layout(xaxis_title='State', yaxis_title='Total Women MSME')
fig_total_msme.show()

# Stacked bar plot of 'Micro', 'Small', and 'Medium' MSMEs by 'State'
fig_msme_types = px.bar(women_df, x='State', y=['Micro', 'Small', 'Medium'], title='Distribution of MSME Types by State',
                       labels={'value': 'Number of MSMEs', 'variable': 'MSME Type'})
fig_msme_types.update_layout(xaxis_title='State', yaxis_title='Number of MSMEs')
fig_msme_types.show()

# Bar plot of 'Employment' by 'State'
fig_employment = px.bar(women_df, x='State', y='Employment', title='Employment Generated by Women-led MSMEs by State')
fig_employment.update_layout(xaxis_title='State', yaxis_title='Employment')
fig_employment.show()

# Bar plot of 'Investment (Cr)' by 'State'
fig_investment = px.bar(women_df, x='State', y='Investment (Cr)', title='Investment in Women-led MSMEs by State')
fig_investment.update_layout(xaxis_title='State', yaxis_title='Investment (Cr)')
fig_investment.show()

# Bar plot of 'Turnover (Cr)' by 'State'
fig_turnover = px.bar(women_df, x='State', y='Turnover (Cr)', title='Turnover from Women-led MSMEs by State')
fig_turnover.update_layout(xaxis_title='State', yaxis_title='Turnover (Cr)')
fig_turnover.show()


# 3. Skills Development Program Analysis
skills2_df = dataframes['skills2']

# Line plot of 'Total Trained (Youth)' over the years
fig_total_trained = px.line(skills2_df, x='Year', y='Total_Trained (Youth)', title='Total Youth Trained Over the Years')
fig_total_trained.update_layout(xaxis_title='Year', yaxis_title='Total Trained (Youth)')
fig_total_trained.show()

# Line plot of 'Total Placed' over the years
fig_total_placed = px.line(skills2_df, x='Year', y='Total_Placed', title='Total Youth Placed Over the Years')
fig_total_placed.update_layout(xaxis_title='Year', yaxis_title='Total Placed')
fig_total_placed.show()

# Stacked area plot of 'Govt_Programs' and 'Private_Programs' over the years
fig_program_types = px.area(skills2_df, x='Year', y=['Govt_Programs', 'Private_Programs'], title='Government vs. Private Skills Programs Over the Years',
                           labels={'value': 'Number of Programs', 'variable': 'Program Type'})
fig_program_types.update_layout(xaxis_title='Year', yaxis_title='Number of Programs')
fig_program_types.show()

# Line plot of 'Budget_Allocated_Cr' over the years
fig_budget = px.line(skills2_df, x='Year', y='Budget_Allocated_Cr', title='Budget Allocated for Skills Development Programs Over the Years')
fig_budget.update_layout(xaxis_title='Year', yaxis_title='Budget Allocated (Cr)')
fig_budget.show()

# Line plot of 'Completion_Rate_pct' over the years
fig_completion_rate = px.line(skills2_df, x='Year', y='Completion_Rate_pct', title='Completion Rate of Skills Development Programs Over the Years')
fig_completion_rate.update_layout(xaxis_title='Year', yaxis_title='Completion Rate (%)')
fig_completion_rate.show()


# 4. IT Companies Revenue Trends
it_companies_df = dataframes['it companies']

# Grouped bar plot of 'Revenue (INR Cr)' by 'Company' and 'FY'
fig_it_revenue = px.bar(it_companies_df, x='Company', y='Revenue (INR Cr)', color='FY',
                       title='Revenue of Major IT Companies by Fiscal Year',
                       barmode='group')
fig_it_revenue.update_layout(xaxis_title='Company', yaxis_title='Revenue (INR Cr)')
fig_it_revenue.show()
```
